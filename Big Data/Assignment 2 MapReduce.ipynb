**Task:** Count number of movies rated by each user

```python
%%file Assignment_2_MapReduce.py
#Simulates the Python command line
#Compare to the command line noted earlier
#!python UserReviews.py u.data
from mrjob.job import MRJob
from mrjob.step import MRStep
import statistics

class UserReviews (MRJob):
    def steps(self):
        return [
            MRStep(mapper=self.mapper_get_movie, 
                  reducer=self.reducer_average_ratings)
        ]

    def mapper_get_movie(self, _, line):
        (user_ID, movieID, rating, timestamp) = line.split('\t')
        ratings = int(rating)
        yield movieID, ratings
        
    def reducer_average_ratings(self, key, ratings):
        yield key, round(statistics.mean(ratings),2)
        
if __name__ == '__main__':
    UserReviews.run()
```
